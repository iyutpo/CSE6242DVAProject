```python
import pandas as pd
```

```python
rt = pd.read_csv('rt.csv')
rt.head()
```

```python
rt = rt.fillna(0)
rt.head()
```

```python
res = pd.DataFrame()

regions = list(rt.region.unique())
dates = list(rt.date.unique())
for i, date in enumerate(dates):
    for j, region in enumerate(regions):
        curr = rt[rt['date'] == date].reset_index(drop=True)
        curr = curr[curr['region'] == region]
        if len(curr) == 0: 
            values = [0] * len(curr.columns)
            values[0], values[1] = date, region
            columns = list(curr.columns.values)
            curr = pd.DataFrame(data=[values], columns=columns)
        res = pd.concat([res, curr])
        
res.head()
```

```python
res.sort_values(by=['date', 'region']).reset_index(drop=True).to_csv('D:\\PSU\\15GA\\MyCourses\\CSE6242DVAProject\\CSE6242DVAProject\\visuals\\data\\rt.csv', index=False)
```
