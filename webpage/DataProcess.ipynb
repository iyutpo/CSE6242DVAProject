```python
import pandas as pd
```

```python
rt = pd.read_csv('./data/combined_data.csv')
rt.head()
```

```python
rt = rt.fillna(0)
rt.head()
```

```python
res = pd.DataFrame()

regions = list(rt.state.unique())
dates = list(rt.date.unique())
for i, date in enumerate(dates):
    for j, region in enumerate(regions):
        curr = rt[rt['date'] == date].reset_index(drop=True)
        curr = curr[curr['state'] == region]
        if len(curr) == 0: 
            values = [0] * len(curr.columns)
            values[0], values[1] = region, date
            columns = list(curr.columns.values)
            curr = pd.DataFrame(data=[values], columns=columns)
        res = pd.concat([res, curr])
res['date'] = pd.to_datetime(res['date'])
res = res.sort_values(by=['date', 'state']).reset_index(drop=True)
res.head()
```

```python
res.to_csv('./data/combined_data1.csv', index=False)
```
